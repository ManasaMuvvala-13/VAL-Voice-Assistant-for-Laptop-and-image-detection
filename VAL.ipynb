```python
import pyjokes
import pyttsx3
import speech_recognition as sr 
import datetime
import wikipedia 
import webbrowser
import os
import smtplib
import numpy as np
import random
import pywhatkit
from email.message import EmailMessage
import ssl

engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)    

def shutdown():
    os.system("shutdown /s /t 1")  

def restart():
    os.system("shutdown /r /t 1")  

def sleep():
    os.system("rundll32.exe powrprof.dll,SetSuspendState 0,1,0")  

def speak(audio):
    engine.say(audio)
    engine.runAndWait()

def wishMe():
    hour = int(datetime.datetime.now().hour)
    if hour>=0 and hour<12:
        speak("Good Morning!")
        print("Good Morning!")

    elif hour>=12 and hour<18:
        speak("Good Afternoon!")   
        print("Good Afternoon!") 

    else:
        speak("Good Evening!")  
        print("Good Evening!")  

    speak("I am VAL. Please tell me how may I help you")

def takeCommand():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        r.pause_threshold = 1
        audio = r.listen(source)

    try:
        print("Recognizing...")    
        query = r.recognize_google(audio, language='en-in')
        print(f"User said: {query}\n")

    except Exception as e:
        print("Say that again please...")  
        return "None"
    return query

def send_email(email_sender, email_password, email_receiver, subject, body):
    em = EmailMessage()
    em['From'] = email_sender
    em['To'] = email_receiver
    em['Subject'] = subject
    em.set_content(body)

    context = ssl.create_default_context()

    with smtplib.SMTP_SSL('smtp.gmail.com', 465, context=context) as smtp:
        smtp.login(email_sender, email_password)
        smtp.sendmail(email_sender, email_receiver, em.as_string())

def listen_command():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        audio = recognizer.listen(source)

    try:
        print("Recognizing...")
        command = recognizer.recognize_google(audio)
        print("You said:", command)
        if "send email" in command:
            print("What is the subject of the email?")
            subject = input_subject()
            print("What is the body of the email?")
            body = input_body()
            send_email(email_sender, email_password, email_receiver, subject, body)
            print("Email sent successfully!")
        else:
            print("Command not recognized.")
    except sr.UnknownValueError:
        print("Could not understand audio.")
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))

def input_subject():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for subject...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        audio = recognizer.listen(source)

    try:
        print("Recognizing subject...")
        subject = recognizer.recognize_google(audio)
        print("Subject:", subject)
        return subject
    except sr.UnknownValueError:
        print("Could not understand audio.")
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))

def input_body():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for body...")
        recognizer.adjust_for_ambient_noise(source, duration=1)
        audio = recognizer.listen(source)

    try:
        print("Recognizing body...")
        body = recognizer.recognize_google(audio)
        print("Body:", body)
        return body
    except sr.UnknownValueError:
        print("Could not understand audio.")
    except sr.RequestError as e:
        print("Could not request results; {0}".format(e))

if __name__ == "__main__":
    email_sender = 'mmanasalahari@gmail.com'
    email_password = 'fypl kgsy uemt jaws'
    email_receiver = 'manasa.20bcd7191@vitap.ac.in'

    wishMe()
    while True:
        query = takeCommand().lower()

        if 'wikipedia' in query:
            try:
                speak('Searching Wikipedia...')
                query = query.replace('wikipedia', '')
                results = wikipedia.summary(query, sentences=2, auto_suggest=False)
                speak('According to Wikipedia')
                print(results.encode('utf-8').decode('utf-8'))  
                speak(results.encode('utf-8').decode('utf-8'))
            except Exception as e:
                print(e)
                speak('Sorry, I could not find any results on Wikipedia.')
                
        elif "stop" in query:
            break

        elif "shutdown" in query:
            shutdown()
            break
        elif "restart" in query:
            restart()
            break
        elif "sleep" in query:
            sleep()
            break

        elif 'open youtube' in query:
            webbrowser.open("youtube.com")
            print("Opening Youtube... ")

        elif 'open google' in query:
            webbrowser.open("google.com")
            print("Opening Google... ")

        elif 'open java point' in query:
            webbrowser.open("https://www.javatpoint.com/")   
            print("Opening Javatpoint")

        elif 'play music' in query:
            song_name = query.replace('play music','')
            speak('The song you asked for is currently playing '+ song_name)
            print(song_name)
            pywhatkit.playonyt(song_name)

        elif 'time' in query:
            strTime = datetime.datetime.now().strftime("%H:%M:%S")    
            speak(f"Ma'am, the time is {strTime}")
            print(strTime)
            

        elif "joke" in query:
            joke_1 = pyjokes.get_joke(language="en",category="neutral")
            print(joke_1)
            speak(joke_1)

        elif 'quit' in query:
            speak("Thank you")
            print("Thank you")
            speak("Have a nice day")
            print("Have a nice day")
            break

        elif "send email" in query:
            print("What is the subject of the email?")
            subject = input_subject()
            print("What is the body of the email?")
            body = input_body()
            send_email(email_sender, email_password, email_receiver, subject, body)
            print("Email sent successfully!")
```
